```python
# Select the variable HP and Attack and calculate the probability distribution

hp_var = pokemon_df['HP']
attack_var = pokemon_df['Attack']

# Determining the Probability distributions
plt.figure(figsize=(8, 5))
sns.kdeplot(hp_var, label='HP')
sns.kdeplot(attack_var, label='Attack')
plt.title('Probability Distribution of HP and Attack')
plt.xlabel('Value')
plt.ylabel('Density')


plt.show # display the plot
```

A KDE plot is a non-parametric way to estimate the probability density function of a random variable. The label parameter is used to set the labels for the hp_var and attack_var variables, and the plt.title(), plt.xlabel(), and plt.ylabel() functions are used to set the title and axis labels for the plot.

Finally, the plt.show() function is used to display the plot.

Note that the skewness value must equal zero for it to be symmetrical, whereas a negative value or less than zero means that it is left-skewed.

## Question No.6:

```python
# Are there any outliers in the variable? 
    # If yes, how are they affecting the distribution?

# Determining IQR for HP
hp_q1 = hp_var.quantile(0.25)
hp_q3 = hp_var.quantile(0.75)
hp_iqr = hp_q3 - hp_q1

# Calculating HP Outlier's lower and upper bounds
hp_lower = hp_q1 - 1.5 * hp_iqr
hp_upper = hp_q3 + 1.5 * hp_iqr

# Removing HP outliers
hp_without_outliers = hp_var[(hp_var >= hp_lower) & (hp_var <= hp_upper)]

# Determining IQR for Attack
attack_q1 = attack_var.quantile(0.25)
attack_q3 = attack_var.quantile(0.75)
attack_iqr = attack_q3 - attack_q1

# Calculating Attack Outlier's lower and upper bounds
attack_lower = attack_q1 - 1.5 * attack_iqr
attack_upper = attack_q3 + 1.5 * attack_iqr

# Removing Attack outliers
attack_without_outliers = attack_var[(attack_var >= attack_lower) & (attack_var <= attack_upper)]

# Plotting the distribution for HP without outliers
sns.displot(hp_without_outliers, color='gold', kde=True)
plt.title('Probability Distribution of HP (Without Outliers)')
plt.xlabel('HP')
plt.ylabel('Probability Density')
plt.show()

# Plotting the distribution for Attack without outliers
sns.displot(attack_without_outliers, color='indigo', kde=True)
plt.title('Probability Distribution of Attack (Without Outliers)')
plt.xlabel('Attack')
plt.ylabel('Probability Density')
plt.show()
```

NOTE: ThiS block of code is used to determine if there are any outliers in the variables hp_var and attack_var.

The code first calculates the interquartile range (IQR) for hp_var and attack_var using the quantile() method from the Pandas library. It then calculates the lower and upper bounds for outliers using the IQR and the formula Q1 - 1.5 * IQR and Q3 + 1.5 * IQR, where Q1 and Q3 are the first and third quartiles, respectively.

The code then removes the outliers from the variables using boolean indexing with the [] operator and the & operator to combine two conditions. The resulting variables without outliers are stored in hp_without_outliers and attack_without_outliers.

Finally, the code plots the probability distribution of hp_without_outliers and attack_without_outliers using the sns.displot() function from the Seaborn library. The color parameter sets the color of the plot, and kde=True adds a kernel density estimate to the plot. The plt.title(), plt.xlabel(), and plt.ylabel() functions set the title and axis labels for the plot.



```python
# Using boxplot to visualize the variables so as to double-check there are no more outliers

# Plotting boxplot for HP without outliers
sns.boxplot(hp_without_outliers, orient='h', color='pink')

# set the title and labels
plt.title("Boxplot for Hp (Without Outliers)")
plt.xlabel("HP")

# show the plot
plt.show()

# Plotting boxplot for Attack
sns.boxplot(attack_without_outliers, orient='h', color='lightblue')

# set the title and labels
plt.title("Boxplot for Attack (Without Outliers)")
plt.xlabel("Attack")

# show the plot
plt.show()
```

ANSWER: Comparing the last visualizations after removing the outliers with the previous ones, it is empirical that there are outliers in the original dataset which affected the probability distribution for HP and Attack variables. After removing the outliers, the visualization now shows a normal distribution that is nearly symmetrical for both HP and Attack variables.
