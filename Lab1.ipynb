Question 4:
Note: The np.random.random function used below is a NumPy function that generates an array of random numbers between 0 and 1 with the specified length. 

```python
# Creating two new variables X and Y with number of values matching length of columns in the imported CSV file

df_Office["X"] = np.random.random(len(df_Office)) #creates a new column named “X” in the DataFrame df_Office and assigns an array of random numbers to it with the same length as the DataFrame.
df_Office["Y"] = np.random.random(len(df_Office)) #creates an array of random numbers with the same length as the DataFrame and assigns it to another new column “Y”.
                                                  #The len(df_Office) function returns the number of elements in the DataFrame.

print(df_Office)
```

Question 5:

```python
# Multiplies X by 100 to scale it to be within 0-100 value range

df_Office["X"] = df_Office["X"]*100 
print(df_Office["X"])
```

```python
# Multiplies Y by 100 to scale it to be within 0-100 value range

df_Office["Y"] = df_Office["Y"]*100
print(df_Office["Y"])
```

Question 6:

```python
# Creating third variable named Z
df_Office["Z"] = df_Office["X"] + df_Office["Y"] # Adds the values of X and Y and assigns the sum to Z.
print(df_Office["Z"])
```

Question 7:

```python
#Creating the forth variable ZZ

df_Office["ZZ"] = np.arange(len(df_Office)) # arranges the number of elements in ascending order starting from zero.
print(df_Office["ZZ"])
```

Question 8:

```python
# Combining X, Y, Z and ZZ to form a table T:

T = np.array ( list (zip (df_Office["X"], df_Office["Y"], df_Office["Z"], df_Office["ZZ"]))) # combines X, Y, Z, and ZZ as a list of array using Numpy, then assings the result to a new variable named T.
print(T)
```

#Question 9:

```python
# Combining T with the original dataset imported from CSV named df_Office

df_newOffice = np.c_[df_Office, T] # concatenates df_Office DataFrame and T using the numpy.c_ function, then assings the result to a new DataFrame named df_newOffice.

# Print the new dataset
print(df_newOffice)
```

#Question 10:

```python
#  Calculating the Mean for Quantiative variables using the np.mean function from the numpy library.
# The [:, i] notation used here means to select all rows and the i-th column from the dataframe.


units_mean = np.mean(df_newOffice[:, 4])
unit_price_mean = np.mean(df_newOffice[:, 5])
X_mean = np.mean(df_newOffice[:, 6])
Y_mean = np.mean(df_newOffice[:, 7])
Z_mean = np.mean(df_newOffice[:, 8])
ZZ_mean = np.mean(df_newOffice[:, 9])

print("The Mean for Units is: ", units_mean)
print("The Mean for Unit Price is: ",unit_price_mean)
print("The Mean for X is: ",X_mean)
print("The Mean for Y is: ",Y_mean)
print("The Mean for Z is: ",Z_mean)
print("The Mean for ZZ is: ",ZZ_mean)
```

```python
#Calculating Median for Quantitative Variables using the np.median function from the numpy library.
# The [:, i] notation used here means to select all rows and the i-th column from the dataframe.

units_median = np.median(df_newOffice[:, 4])
unit_price_median = np.median(df_newOffice[:, 5])
X_median = np.median(df_newOffice[:, 6])
Y_median = np.median(df_newOffice[:, 7])
Z_median = np.median(df_newOffice[:, 8])
ZZ_median = np.median(df_newOffice[:, 9])
```
